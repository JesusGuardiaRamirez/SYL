```python
import sounddevice as sd
import soundfile as sf
import speech_recognition as sr
import PyAudio

# Configuración de parámetros
duration = 30  # Duración de la grabación en segundos
sample_rate = 44100  # Tasa de muestreo en Hz
channels = 2  # Número de canales de audio (estéreo)


# Definir la palabra clave que se debe detectar
palabra_clave = "activar"

# Inicializar el reconocedor de voz
recognizer = sr.Recognizer()

while True:
    # Capturar el audio del micrófono
    with sr.Microphone() as source:
        print("Escuchando...")
        audio = recognizer.listen(source)

    try:
        # Reconocer el texto del audio
        texto = recognizer.recognize_google(audio, language="es-ES")
        print("Texto reconocido:", texto)

        # Verificar si la palabra clave está presente en el texto
        if palabra_clave in texto.lower():
            print("Palabra clave detectada. Activando el código...")
            # Aquí puedes colocar el código que deseas ejecutar cuando se detecte la palabra clave
            
        # Iniciar la grabación de audio desde el dispositivo de entrada predeterminado
        recording = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=channels)

        print("Grabando")

        try:
        #Esperar a que finalice la grabación
            sd.wait()
        except KeyboardInterrupt:
            pass

        # Guardar los datos grabados en un archivo WAV
        filename = "conversacion.wav"
        sf.write(filename, recording, sample_rate)

        # Cargar el archivo WAV
        audio_file = sr.AudioFile(filename)

        # Convertir el audio a texto usando la API de reconocimiento de voz de Google
        recognizer = sr.Recognizer()
        with audio_file as source:
            audio_data = recognizer.record(source)
            try:
                text = recognizer.recognize_google(audio_data, language='es-ES')
            except sr.UnknownValueError:
                text = "No se pudo reconocer ningún texto en el audio grabado."

        # Mostrar la conversación escrita por consola
        print("Nombre:")
        print("Ubicación:")
        print("Conversación escrita:")
        print(text)

        # Guardar la conversación escrita en un archivo de texto
        ruta_archivo = "/Users/JGR/Desktop/reconocimiento_voz/text.txt"
        with open(ruta_archivo, 'w') as file:
            file.write(text)
                    
        break  # Romper el ciclo para detener la escucha después de detectar la palabra clave

    except sr.UnknownValueError:
        print("No se pudo reconocer el audio.")
    except sr.RequestError as e:
        print("Error al solicitar el servicio de reconocimiento de voz:", str(e))

```
